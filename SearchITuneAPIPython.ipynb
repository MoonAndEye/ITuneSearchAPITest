```python
import urllib
import json
```

```python
host = "https://itunes.apple.com/search"

#如果要換歌手，就換下面這個
term = "?term=周杰倫"

url = host + term

result = urllib.urlopen(url)

responseJson = json.loads(result.read())

responseJson.get("results")

resultDict = responseJson.get("results")[0]

resultDict["previewUrl"]
```
